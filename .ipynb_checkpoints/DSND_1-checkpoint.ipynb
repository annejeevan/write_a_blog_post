#### Missing value distribution

```python
def get_missing_percentage(series):
    """ Calculates percentage of NaN values in DataFrame
    :param series: Pandas DataFrame object
    :return: float
    """
    num = series.isnull().sum()
    den = len(series)
    return round(num/den, 2)
```

##### Imputation

```python
#imputing the missing values with the median, as they are robust to outliers
#Impute the values using scikit-learn SimpleImpute Class
price_cal_med = features_df['price_cal'].median()
security_deposit_med = features_df['security_deposit'].median()
cleaning_fee_med = features_df['cleaning_fee'].median()

features_df['security_deposit'] = features_df['security_deposit'].fillna(security_deposit_med)
features_df['cleaning_fee'] = features_df['cleaning_fee'].fillna(cleaning_fee_med)
features_df['price_cal'] = features_df['price_cal'].fillna(price_cal_med)
```

```python
# Only include columns that contain any NaN values
features_df_any_null = features_df[features_df.columns[features_df.isnull().any()].tolist()]
# see if there are still missing values exists
get_missing_percentage(features_df_any_null)
```

### Whats factors contribute to the better rating?

Lets build a machine learning model and see the which features are good for rating

```python
# Labels are the values we want to predict
labels = np.array(features_df['review_scores_rating'])

# Remove the labels from the features
# axis 1 refers to the columns
features_df= features_df.drop('review_scores_rating', axis = 1)

# Saving feature names for later use
feature_list = list(features_df.columns)

# Convert to numpy array
features = np.array(features_df)
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

```python
# number of training and testing observations
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

#### Linear Regression

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import (LinearRegression, Ridge, 
                                  Lasso, RandomizedLasso)
from sklearn.feature_selection import RFE, f_regression
from sklearn.preprocessing import MinMaxScaler
import numpy as np
    
```

```python
names = ["%s" % i for i in feature_list]
 
ranks = {}
 
def rank_to_dict(ranks, names, order=1):
    """
    arrange in a dictionary
    """
    minmax = MinMaxScaler()
    ranks = minmax.fit_transform(order*np.array([ranks]).T).T[0]
    ranks = map(lambda x: round(x, 2), ranks)
    return dict(zip(names, ranks ))

## lasso regression
lasso = Lasso(alpha=.05)
lasso.fit(train_features, train_labels)
ranks["Lasso"] = rank_to_dict(np.abs(lasso.coef_), names)
```

```python
pd.DataFrame.from_dict(ranks['Lasso'], orient='index',columns=['coef']).sort_values('coef',ascending=False)[:15]
```

From the model we created to predict rating, the contributing factors are:

- People usually see the reviews of the listings and then book, as we see in our model that increase in review scores accuracy, value, cleanliness, communication, checkin increases the review ratings
- Also, review might increase when host is superhost. To be a superhost, need to be good in all aspects (service, cleanliness etc)

In order to get reviews, service should be better for the value of the listing.
